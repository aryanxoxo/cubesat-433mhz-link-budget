# 433 MHz CubeSat Link Budget and BER Model

Analysis notebook for the UBC ORBIT / ALEASAT 433 MHz CubeSat radio link. The values are editable reference assumptions used to demonstrate the calculation workflow before substituting measured hardware and station parameters.

## Equations

Free-space path loss:

```text
FSPL(dB) = 20log10(range_km) + 20log10(frequency_MHz) + 32.44
```

Received power:

```text
Pr(dBm) = Ptx + Gtx + Grx - FSPL - losses
```

Noise floor:

```text
N(dBm) = -174 + 10log10(BW_Hz) + NF
```
